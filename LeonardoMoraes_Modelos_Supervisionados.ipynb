```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
import seaborn as sns
```

```python
# Carregando a base
df_close = pd.read_csv('/content/housing_treino.csv')
```

```python
df_close
```

```python
# Usando StatsModels para montar um modelo de regressão linear simples
rls = sm.ols(formula = 'Sale_Price ~ Gr_Liv_Area', data = df_close).fit()
print(rls.summary())
```

```python
# Montando um modelo de regressão linear múltipla
rls_a = sm.ols(formula = 'Sale_Price ~ Gr_Liv_Area+Lot_Area', data = df_close).fit()
print(rls_a.summary())
```

```python
# Montando um modelo de regressão linear múltipla
rls_b = sm.ols(formula = 'Sale_Price ~ Gr_Liv_Area+Lot_Area+Bedroom_AbvGr', data = df_close).fit()
print(rls_b.summary())
```

```python
# Montando um modelo de regressão linear múltipla
rls_c = sm.ols(formula = 'Sale_Price ~ Gr_Liv_Area+Lot_Area+Bedroom_AbvGr+Lot_Frontage', data = df_close).fit()
print(rls_c.summary())
```

```python
# Montando um modelo de regressão linear múltipla
rls_d = sm.ols(formula = 'Sale_Price ~ Gr_Liv_Area+Lot_Area+Bedroom_AbvGr+Lot_Frontage+Garage_Area', data = df_close).fit()
print(rls_d.summary())
```

```python
# Montando DF com as previsões
resultado = pd.DataFrame({'Sale_Price': df_close['Sale_Price'],
                          'RLS': rls.predict(),
                          'RL_2': rls_a.predict(),
                          'RL_3': rls_b.predict(),
                          'RL_4': rls_c.predict(),
                          'RL_ALL': rls_d.predict()},
                          index = df_close.index
                          )
```

```python
# Visualizando as previsões para comparação
fig, ax = plt.subplots(figsize=(20,10))
sns.lineplot(data=resultado)
```

```python
# Análise de resíduos do melhor modelo
rls_b.resid.plot.density()
```

Sim, o uso da abordagem de treino-teste é fundamental em problemas de Machine Learning, já que um conjunto de testes independentes permita que seja verificada a eficácia de seu modelo o que é importante para entender se o modelo está apenas memorizando o conjunto de treinamento (overfitting) ou não generalizando bem a partir dele (underfitting). Além disso, existem as questões da previsão da performance do modelo já que a divisão treino-teste é capaz de fornecer uma indicação de desempenho e a melhoria da robustez, já que podemos usar os resultados dos testes para ajustar o modelo.

```python
# Carregando a base
data_1 = pd.read_excel('/content/mtcars_1.xls')
```

```python
# Criando o objeto StandardScaler
scaler = StandardScaler()
```

```python
# Separando as variáveis independentes e a variável dependente
X = data_1.drop('vs', axis=1)  # X inclui todas as colunas exceto 'vs'
y = data_1['vs']  # y é a variável dependente 'vs'
```

```python
# Padronizando os dados
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)  # convertendo novamente para DataFrame
```

```python
# Dividindo os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Criando o objeto de regressão logística
logistic_model = LogisticRegression()
```

```python
# Treinando o modelo usando os dados de treinamento
logistic_model.fit(X_train, y_train)
```

```python
# Fazendo previsões nos dados de teste
y_pred = logistic_model.predict(X_test)
```

```python
# Calculando a acurácia
accuracy = accuracy_score(y_test, y_pred)
print(f'Acurácia do modelo: {accuracy*100:.2f}%')
```

```python
# Matriz de confusão
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

```python
# Classification Report
print(classification_report(y_test, y_pred))
```

Acurácia de 86%. O seu TN foi de 2 em 3 ocorrências da classe-alvo (em "1").

```python

# Calculando as probabilidades das previsões
y_pred_proba = logistic_model.predict_proba(X_test)[:,1]

# Calculando a curva ROC
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)

# Calculando a área sob a curva (AUC)
roc_auc = roc_auc_score(y_test, y_pred_proba)

# Plotando a curva ROC
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
# Carregando  as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Carregando a base
data_2 = pd.read_excel('/content/mtcars_1.xls')
```

```python
data_2
```

```python
# Separando o dataset em variáveis independentes e variável-alvo
feature_cols = ['mpg', 'cyl', 'disp', 'hp','wt']
X = data_2[feature_cols] # Variáveis independentes
y = data_2.vs # Variável-alvo
```

```python
# Dividindo a base em conjuntos de Treino e Teste (70% Treino and 30% Teste)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Instanciando, treinando o modelo e aplicando o resultado na base de testes

# Instanciando o classificador
clf = DecisionTreeClassifier()

# Realizando o treino
clf = clf.fit(X_train,y_train)

# Aplicando o modelo treinado no conjunto de Teste
y_pred = pd.DataFrame(clf.predict(X_test))
```

```python
# Avaliando os resultados

# Matriz de confusão
# Linha: Real
# Coluna: Predito
print(confusion_matrix(y_test, y_pred))
```

```python
# Classification Report
print(classification_report(y_test, y_pred))
```

Acurácia de 100%. A matriz confusão indica o acerto de todas as ocorrências da classe-alvo.

```python
# Visualização da árvore gerada.
from sklearn import tree

fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (50,50), dpi=300)
tree.plot_tree(clf, filled=True)
fig.savefig('tree_sklearn.png')
```

No caso da regressão logística, o modelo apresentou uma acurácia de 86%, já que o seu TN identificou 2 ocorrências da variável-alvo (valor “1”) de 3 e o modelo previu incorretamente a classe negativa como positiva. No caso da árvore de decisão a matriz confusão indica o acerto de todas as ocorrências da classe-alvo, com uma acurácia de 100%. A árvore de decisão apresentou uma performance melhor para o exercício em questão.

```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
```

```python
# Fazendo a separação dos dados (variáveis independente e dependente)
X = data_3[[
    'RowNumber',
    'CustomerId',
    'CreditScore',
    'Age',
    'Tenure',
    'Balance',
    'NumOfProducts',
    'HasCrCard',
    'IsActiveMember',
    'EstimatedSalary',
]]


y = data_3['Exited']
```

```python
# Garantindo que não existem NAs para não termos problemas na modelagem.
X = X.fillna(0)

# Transformando tudo para float
X = X.astype(float)
```

```python
# Separando dados para treino e teste
# São dois grupos de dados, um para uso no modelo e outro para uso na validação.

# Separando os dados de treino e teste para o modelo
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Normalizando as variáveis
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Instanciando o modelo
clf = MLPClassifier(hidden_layer_sizes=(10, 10, 10),
                    max_iter=1000,
                    activation = 'logistic',
                    solver='adam'
                   )
clf.fit(X_train,y_train)

# Avaliando uma única execução
y_pred_teste = clf.predict(X_test)
```

```python
print(metrics.confusion_matrix(y_test, y_pred_teste))
print(metrics.classification_report(y_test, y_pred_teste,
                                    target_names=['Saudavel','Suspeito']))
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_teste))
print("Precision:",metrics.precision_score(y_test, y_pred_teste))
print("Recall:",metrics.recall_score(y_test, y_pred_teste))
```

```python
# Instanciando o modelo
clf = MLPClassifier(hidden_layer_sizes=(5, 5, 5),
                    max_iter=1000,
                    activation = 'logistic',
                    solver='adam'
                   )
clf.fit(X_train,y_train)

# Avaliando uma única execução
y_pred_teste = clf.predict(X_test)
```

```python
print(metrics.confusion_matrix(y_test, y_pred_teste))
print(metrics.classification_report(y_test, y_pred_teste,
                                    target_names=['Saudavel','Suspeito']))
print("Accuracy:",metrics.accuracy_score(y_test, y_pred_teste))
print("Precision:",metrics.precision_score(y_test, y_pred_teste))
print("Recall:",metrics.recall_score(y_test, y_pred_teste))
```
